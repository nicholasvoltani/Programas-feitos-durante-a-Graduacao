   # Lorenz Attractor

```python
from solvers import Euler, RK4
import matplotlib.pyplot as plt
import numpy as np
```

The differential equations system of a Lorenz attractor is given by:

$$\begin{align}
    \begin{cases}
    \frac{dx}{dt}&= \sigma (y-x)\\
    \frac{dy}{dt}&= x(\rho - z) - y\\
    \frac{dz}{dt}&= xy - \beta z
    \end{cases}
\end{align}$$

```python
x=[-15.1]
y=[-15.1]
z=[13.1]
params = [8/3,28,10] ## \beta, \rho, \sigma
def fx(t,x,y,z,b,r,s):
    return s*(y-x)
def fy(t,x,y,z,b,r,s):
    return x*(r-z) - y
def fz(t,x,y,z,b,r,s):
    return x*y - b*z
funcs = [fx,fy,fz]
T = 100
dt = 0.01
for t in range(int(T/dt)):
    pos = np.array([x[t],y[t],z[t]])
    pos = RK4(t,dt,pos,params,funcs)
    x.append(pos[0])
    y.append(pos[1])
    z.append(pos[2])
    
fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection = '3d')
ax.scatter3D(x,y,z,s=5,c = plt.cm.hsv(np.linspace(0,T,int(T/dt)+1)))
plt.show()
```
